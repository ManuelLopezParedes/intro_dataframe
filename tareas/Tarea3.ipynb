```python
# importamos librerias
import pandas as pd
import numpy as np
```

```python
# cargamos el DF
df = pd.read_csv('empleados.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# modificar cabecera
df.columns=['nombre','sexo','fecha_inicio','ultimo_inicio','salario','bonificacion','alta_direccion','equipo']
df.columns
```

```python
df.sample(2)
```

```python
df['alta_direccion'] = df['alta_direccion'].astype("bool")
df['sexo'] = df['sexo'].astype("category")
df['salario'] = df['salario'].astype("int64")
df['bonificacion'] = df['bonificacion'].astype("float64")
df['ultimo_inicio'] = pd.to_datetime(df['ultimo_inicio'], format='%I:%M %p').dt.time
df['fecha_inicio'] = pd.to_datetime(df['fecha_inicio'], format='%m/%d/%Y')
df.info()
```

```python
df.sample(3)
```

```python
df = df.dropna(subset=["nombre","sexo","equipo"]).reset_index(drop=True)
df.info()
```
