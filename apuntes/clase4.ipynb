# Filtrar informacion de u DF

```python
df['Hora del último inicio de sesión'] = pd.to_datetime(df['Hora del último inicio de sesión'],format="%H:%M %p").dt.time
```

```python
df.info()
```

```python
df['Alta dirección'] = df['Alta dirección'].astype("bool")
df['Alta dirección']
```

```python
df['Sexo'] = df['Sexo'].astype("category")
df['Sexo']
```

#Filtrado por serie
obtener los registros que son del equipo de marketing


```python
resultado =df['Equipo'] == 'Marketing'
# si imprimimos resultado nos dara una lista de todos los trabajadores
resultado
```

```python
# Para mostrar los registros usamos el siguiente codigo
df[resultado]
# ó
df[df['Equipo']=='Marketing']
```

### Filtrar solo a los que ganen mas de 45000

```python
df[df['Salario']>45000]
```

### Filtar los que se contrataron antes de 2/27/2000

```python
df[pd.to_datetime(df['Fecha de inicio']) < pd.to_datetime('2000-02-27')]
```

### Empleados que hayan iniciado sesion antes de las 7:00AM

```python
df[df['Hora del último inicio de sesión'] < pd.to_datetime('07:00:00').time()]
```

### Filtrar los valores numericos mas altos o mas bajos

```python
# top 3 salarios
df.nlargest(n=3,columns="Salario")
```

```python
# top 5 salarios mas bajos
df.nsmallest(n=5,columns="Salario")
```

# Operadores booleanos en filtrado de DF

## AND &

```python
gender =df['Sexo'] == "Female"
depto = df['Equipo'] =="Finance"

df[gender & depto]
```

```python
# Filtar a los empleados que sean del equipo legal y que su bonificacion sea menor a 6%
depto = df['Equipo']=="Legal"
bonificacion = df['bonificación %'] < 6

df[depto & bonificacion]
```

## OR |

```python
# filtrar a los empleados hombres o que son del equipo producto o que ganen < 45000
sexo = df['Sexo'] == "Hombre"
depto = df['Equipo'] == "Product"
salario = df['Salario'] < 45000

df[sexo | depto | salario]
```

```python
df[((df['Sexo'] == 'Female') & ((df['Equipo'] == 'Sales') | (df['Equipo'] == 'Legal')))]
```

## IN

```python
# a los empleados de nombre Justin, Susan o Ann
df[df['Nombre'].isin(['Justin','Susan','Ann'])]
```

```python
#selecciona a los empleados que ganen 38344 o 71958
df[df['Salario'].isin([38344, 71958])]
```

## Emplear operadores logicos con isna y notnull

```python
df[df["Sexo"].isnull() & df["Equipo"].isnull()]
```

## Between

```python
df[df["Salario"].between(90000,95000)]
```

```python
# ORDER By
df[df['Salario'].between(94000,95000)].sort_values(by='Salario')
```

```python
df.sort_values(by="bonificación %", ascending=False)
```

## Duplicados

```python
df[df["Nombre"].duplicated()]
```
